## iris data

This notebook provides a simple example of clustering with sklearn. It runs the kmeans algorithm on the iris data.

```python
# imports used in this notebook

import pandas as pd
import numpy as np
import sklearn.metrics as sm
from sklearn.cluster import KMeans
from sklearn import datasets
import seaborn as sb
```

```python
### load the data

iris = datasets.load_iris()
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data)
df.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_width']
df.head()
```

```python
model=KMeans(n_clusters=3)
model.fit(df)
```

```python
model.labels_
```

```python
# plot actual labels using seaborn

sb.scatterplot(x='Sepal_Length', y='Sepal_width', hue=iris.target, data=df)
```

```python
# plot cluster results using seaborn

sb.scatterplot(x='Sepal_Length', y='Sepal_width', hue=model.labels_, data=df)

```
