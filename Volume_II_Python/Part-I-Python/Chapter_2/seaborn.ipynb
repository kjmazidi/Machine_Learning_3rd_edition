```python
# imports used in this notebook

import pandas as pd
import seaborn as sb
from sklearn import datasets
```

```python
# load iris
iris = datasets.load_iris()
X = iris.data
y = iris.target

# convert to data frame to make seaborn happy
df = pd.DataFrame(X, columns=iris.feature_names)
df_y = pd.DataFrame(y, columns=['species'])
```

### Plot a quantitative array

The code below plots the petal length column of the data frame. The parameter kde=True plots the curve on top of the distribution. The parameter rug=True makes tick marks for observations across the bottom of the chart.

```python
sb.displot(df['petal length (cm)'], kde=True, rug=True)
```

### Plot two quantitative arrays

Plotting quantitative arrays on the x and y axes shows the relationship between the two variables. In the code below, both the color and shape of the points is altered according to the class, y.

```python
sb.relplot(x='petal length (cm)', y='petal width (cm)', data=df, hue=df_y.species, style=df_y.species)
```

### Plotting categorical data

The target column, species, is categorical data. Species can take on one of three values. In this data set, species is equally distributed among the 3 classes, with 50 observations in each class.

```python
sb.catplot(x="species", kind='count', data=df_y)
```

### Boxplot example

```python
# add the species column to the data frame to make the graph syntax simpler

df['species'] = df_y
sb.boxplot(data=df, x='species', y='petal width (cm)')
```
